#### Update Nan values in Manufacturer based on model

- Create a Mapping from model to manufacturer: Use the existing rows where both model and manufacturer are present to create a dictionary mapping.
- Fill Missing manufacturer Values: Use this mapping to fill in the missing manufacturer values in the DataFrame.

```python
# Create a mapping from model to manufacturer using non-null values
model_to_manufacturer = cars_cleaned.dropna(subset=['manufacturer', 'model']).set_index('model')['manufacturer'].to_dict()

# Function to fill missing manufacturer based on model
def fill_manufacturer(row, mapping):
    if pd.isnull(row['manufacturer']) and pd.notnull(row['model']):
        return mapping.get(row['model'], row['manufacturer'])
    return row['manufacturer']

# Apply the function to each row in the DataFrame
cars_cleaned['manufacturer'] = cars_cleaned.apply(fill_manufacturer, axis=1, mapping=model_to_manufacturer)

cars_cleaned[cars_cleaned['manufacturer'].isnull()].shape
```

```python
# Drop rows with missing 'model' values
cars_cleaned.dropna(subset=['model'], inplace=True)
```

```python
# Display counts of missing values
cars_cleaned.isnull().sum()
```

**Add a New Feature - make_model (combination of manufacturer and model)**

```python
# Create 'make_model' column based on the conditions
cars_cleaned['make_model'] = np.where(
    cars_cleaned['manufacturer'].isnull() & cars_cleaned['model'].notnull(), 
    cars_cleaned['model'],
    np.where(
        cars_cleaned['model'].isnull() & cars_cleaned['manufacturer'].notnull(), 
        cars_cleaned['manufacturer'], 
        np.where(
            cars_cleaned['manufacturer'].notnull() & cars_cleaned['model'].notnull(), 
            cars_cleaned['manufacturer'] + '-' + cars_cleaned['model'],
            np.nan
        )
    )
)
```

#### Outlier Detection
Identified outliers in numerical features by visualizing box plots and scatter plots. Assessed the potential impact of these outliers on the analysis and modeling process.

<b> Handling Outliers:</b> replace with mode.

```python
# Function to identify outliers using IQR
def identify_outliers(df, feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
    return outliers

# Identify outliers in each feature
outliers_price = identify_outliers(cars_cleaned, 'price')
outliers_year = identify_outliers(cars_cleaned, 'year')
outliers_odometer = identify_outliers(cars_cleaned, 'odometer')

print("Outliers in Price:\n", outliers_price)
print("Outliers in Year:\n", outliers_year)
print("Outliers in Odometer:\n", outliers_odometer)
```

```python
# Handle outliers by replacing with mode
def handle_outliers(df, feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    mode = df[feature].mode()[0]
    df[feature] = df[feature].fillna(mode)
    df[feature] = df[feature].apply(lambda x: mode if x < lower_bound or x > upper_bound else x)
    return df

# Clean the dataset by removing outliers
for feature in ['price', 'year', 'odometer']:
    cars_cleaned = handle_outliers(cars_cleaned, feature)
```

# Feature Engineering
This process involves selecting, adjusting, and converting raw data into features suitable for supervised learning.

## Preprocessing Steps

1. **Handling Missing Data**:
   - For the numerical fields such as `vehicle_age` and `odometer`, missing values were addressed with `SimpleImputer(strategy='median')`. Fortunately, these specific variables had no missing entries, ensuring their completeness.

2. **Standardization**:
   - The numerical variables (`vehicle_age` and `odometer`) were standardized using the `StandardScaler` to maintain uniformity across the dataset.

3. **Polynomial Feature Creation**:
   - A polynomial transformation of degree 2 was applied to the numerical variables (`vehicle_age` and `odometer`) to capture potential non-linear relationships within the data.

4. **Imputation for Categorical Variables**:
   - Missing values in categorical features were filled using `SimpleImputer(strategy='most_frequent')`, considering the specific make and model associated with each entry.

5. **Target Encoding for Categorical Features**:
   - The `make_model` categorical variable was transformed using Target Encoding, which replaces categories with the average target value linked to them.

6. **Ordinal Encoding**:
   - Certain categorical variables, including `condition`, `cylinders`, `title_status`, `drive`, and `size`, were transformed through ordinal encoding, enabling meaningful numerical representation.

7. **One-Hot Encoding**:
   - Categorical features like `fuel`, `transmission`, `type`, and `paint_color` were converted into binary format via one-hot encoding, creating distinct columns for each category.

8. **Passing Through Unmodified Columns**:
   - Any remaining columns not specifically processed were retained as they were by setting `remainder='passthrough'`.

This preprocessing approach ensured that categorical data could be effectively integrated into predictive models while avoiding unnecessary ordinality.

## Target Encoding for Categorical Features (make_model)
Target Encoding was utilized for categorical variables, where values were substituted with the average target value corresponding to `make_model`.

### Benefits:
- **Information Retention**: This technique preserves the unique details of each manufacturer and model in a numerical format, reflecting their distinct characteristics.
  
- **Trend Identification**: By encoding categories with their average prices, this method captures essential pricing trends related to each manufacturer and model, which is critical for price prediction.

- **Efficient Handling of High Cardinality**: Given that manufacturers and models typically exhibit high cardinality (many unique entries), Target Encoding effectively manages this without generating excessive additional columns, unlike One-Hot Encoding.


```python
# Columns to be target encoded
target_enc_features = ['make_model']

# Initialize TargetEncoder
te = TargetEncoder(cols=target_enc_features)

# Fit and transform on training data
X_train_te = te.fit_transform(X_train[target_enc_features], y_train)
X_train_te = X_train_te.sort_index()

# Transform test data using fitted encoder
X_test_te = te.transform(X_test[target_enc_features]).sort_index()
X_train_te = X_train_te.sort_index()
X_train_te
```

```python
# Drop original columns and concatenate the encoded columns
X_train_combined = X_train.drop(columns=target_enc_features).sort_index()
X_test_combined = X_test.drop(columns=target_enc_features).sort_index()

# Concatenate encoded columns
X_train_combined = pd.concat([X_train_combined, X_train_te], axis=1).sort_index()
X_test_combined = pd.concat([X_test_combined, X_test_te], axis=1).sort_index()
X_train_combined.shape, X_train_combined.info()
```

### Considerations for Target Encoding:
- **Data Quantity and Distribution**: It is essential to have a sufficient number of data points for each manufacturer and model to compute accurate average prices. Limited sample sizes can result in skewed estimates.

- **Validation Approach**: To prevent data leakage when employing Target Encoding, average prices must be calculated solely from the training data. The encoder should be fitted on the training set, ensuring that both the training and testing datasets are transformed uniformly.

- **Effect on Model Efficiency**: Assess whether the encoded features enhance model performance through techniques like cross-validation or by using a separate validation dataset. In some cases, more straightforward encoding methods may be adequate based on the characteristics of the dataset and the chosen model.

### Ordinal Encoding for Categorical Features
Certain ordinal categorical features such as `condition`, `cylinders`, `title_status`, `drive`, and `size` were encoded using the `OrdinalEncoder`, following a specified order of categories.

#### Categories for Encoding:
- **Condition**: Salvage, Fair, Good, Excellent, Like New, New

- **Cylinders**: Other, 3 Cylinders, 4 Cylinders, 5 Cylinders, 6 Cylinders, 8 Cylinders, 10 Cylinders, 12 Cylinders

- **Title Status**:
  - **Missing**: Title is absent; the vehicle cannot be sold or registered legally.
  - **Parts Only**: The vehicle is only suitable for parts and cannot be legally driven.
  - **Salvage**: The vehicle was damaged and declared a total loss by an insurance company.
  - **Rebuilt**: The vehicle was previously salvaged but has been repaired and certified.
  - **Lien**: The vehicle is under a lien, meaning it serves as collateral for a loan.
  - **Clean**: The vehicle has a clean title, indicating no major accidents or significant damage.

- **Drive Type**:
  - **RWD (Rear-Wheel Drive)**: Generally perceived as less adaptable in poor weather conditions.
  - **FWD (Front-Wheel Drive)**: More prevalent in consumer vehicles and performs better in slippery conditions compared to RWD.
  - **4WD (Four-Wheel Drive)**: Usually regarded as the most versatile, suited for off-road and all-weather scenarios.

- **Size Category**:
  - **Sub-Compact**: The smallest category, typically very compact and economical.
  - **Compact**: Slightly larger than sub-compact, yet still smaller than mid-size.
  - **Mid-Size**: An intermediate size category, offering a balance between compactness and spaciousness.
  - **Full-Size**: The largest category, providing maximum space and comfort.


```python
# Define ordinal encoding pipeline
ordinal_features = ['condition', 'cylinders', 'title_status', 'drive', 'size']

# Define the order of categories for the 'condition' and 'cylinders' variables
condition_categories = ['salvage', 'fair', 'good', 'excellent', 'like new', 'new']
cylinders_categories = ['other', '3 cylinders', '4 cylinders', '5 cylinders', '6 cylinders', '8 cylinders', '10 cylinders', '12 cylinders']
title_categories = ['missing', 'parts only', 'salvage', 'rebuilt', 'lien', 'clean']
drive_categories = ['rwd', 'fwd', '4wd']
size_categories = ['sub-compact', 'compact', 'mid-size', 'full-size']

# Define ordinal encoding transformers for each categorical variable with list of indices
ordinal_transformers = [
    ('condition', OrdinalEncoder(categories=[condition_categories]), ['condition']),
    ('cylinders', OrdinalEncoder(categories=[cylinders_categories]), ['cylinders']),
    ('title_status', OrdinalEncoder(categories=[title_categories]), ['title_status']),
    ('drive', OrdinalEncoder(categories=[drive_categories]), ['drive']),
    ('size', OrdinalEncoder(categories=[size_categories]), ['size'])
]

# Create the ordinal encoding pipeline using ColumnTransformer
ordinal_pipeline = Pipeline(steps=[
    ('ordinal', ColumnTransformer(transformers=ordinal_transformers))
])
```

```python
# Fit and transform on training data
X_train_ord = ordinal_pipeline.fit_transform(X_train_combined, y_train)
X_test_ord = ordinal_pipeline.transform(X_test_combined)

X_train_ord = pd.DataFrame(X_train_ord, columns=['condition_ord', 'cylinders_ord', 'title_status_ord', 'drive_ord', 'size_ord'])
X_train_ord.index = X_train_combined.index
X_test_ord = pd.DataFrame(X_test_ord, columns=['condition_ord', 'cylinders_ord', 'title_status_ord', 'drive_ord', 'size_ord'])
X_test_ord.index = X_test_combined.index
```

#### **One Hot Encoder for Categorical Variables (fuel, transmission, type, paint_color)**
- Encode nominal categorical features using OneHotEncoder, ignoring unknown categories.

```python
# Define one hot encoder pipeline
ohe_features = ['fuel', 'transmission', 'type', 'paint_color']

# Define categorical one-hot encoding pipeline with imputer
# Create the one-hot encoding pipeline
one_hot_pipeline = Pipeline(steps=[
    ('onehot', ColumnTransformer(
        transformers=[
            ('onehot', OneHotEncoder(), ohe_features)
        ]
    ))
])

# Fit and transform on training data
X_train_te_oe_ohe = one_hot_pipeline.fit_transform(X_train_te_oe[ohe_features], y_train)
X_test_te_oe_ohe = one_hot_pipeline.transform(X_test_te_oe[ohe_features])

# Check the shape of the transformed array
num_columns = X_train_te_oe_ohe.shape[1]
print(f"Total number of columns after one-hot encoding: {num_columns}")

# Get the feature names from the OneHotEncoder
ohe_feature_names = one_hot_pipeline.named_steps['onehot'].named_transformers_['onehot'].get_feature_names_out(ohe_features)
X_train_te_oe_ohe_df = pd.DataFrame(X_train_te_oe_ohe.toarray() if hasattr(X_train_te_oe_ohe, "toarray") else X_train_te_oe_ohe, columns=ohe_feature_names)
X_test_te_oe_ohe_df = pd.DataFrame(X_test_te_oe_ohe.toarray() if hasattr(X_test_te_oe_ohe, "toarray") else X_test_te_oe_ohe, columns=ohe_feature_names)

X_train_te_oe_ohe_df.index = X_train_te_oe.index
X_test_te_oe_ohe_df.index = X_test_te_oe.index

# Drop original columns and concatenate the encoded columns
X_train_txd = X_train_te_oe.drop(columns=ohe_features).sort_index()
X_test_txd = X_test_te_oe.drop(columns=ohe_features).sort_index()

# Concatenate encoded columns
X_train_txd = pd.concat([X_train_txd, X_train_te_oe_ohe_df], axis=1).sort_index()
X_test_txd = pd.concat([X_test_txd, X_test_te_oe_ohe_df], axis=1).sort_index()

X_train_txd
```

```python
from sklearn.pipeline import Pipeline

poly_features = ['odometer', 'make_model']

# Create the pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),            # Step 1: Scale the features
    ('poly', PolynomialFeatures(degree=2, include_bias=False))  # Step 2: Create polynomial features
])
X_train_xd = pipeline.fit_transform(X_train_txd[poly_features], y_train)
X_test_xd = pipeline.transform(X_test_txd[poly_features])

# Get feature names from polynomial transformation
poly_feature_names = pipeline.named_steps['poly'].get_feature_names_out(input_features=poly_features)

# Convert transformed data to DataFrame with correct column names
X_train_xd_df = pd.DataFrame(X_train_xd, columns=poly_feature_names)
X_test_xd_df = pd.DataFrame(X_test_xd, columns=poly_feature_names)
X_train_xd_df.index = X_train_txd.index
X_test_xd_df.index = X_test_txd.index

# Drop original columns and concatenate the encoded columns
X_train_fx = X_train_txd.drop(columns=poly_features).sort_index()
X_test_fx = X_test_txd.drop(columns=poly_features).sort_index()
# Concatenate encoded columns
X_train_fx = pd.concat([X_train_fx, X_train_xd_df], axis=1).sort_index()
X_test_fx = pd.concat([X_test_fx, X_test_xd_df], axis=1).sort_index()
```

### Evaluation

In the Evaluation phase, the trained models were assessed to verify that they aligned with project goals and business requirements:

**Evaluation Metrics:** Metrics like Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R-squared (R²) were used to evaluate the prediction accuracy. These metrics helped gauge how closely the models' predicted used car prices matched the actual values.

**Model Validation:** The models were validated on the test dataset, reserved during data preparation, to ensure they generalized well to new, unseen data and were not overfitting to the training data.

**VIF for Multicollinearity:** Variance Inflation Factor (VIF) was calculated to detect multicollinearity among the independent variables. VIF quantifies how much the variance of a regression coefficient increases due to collinearity with other features. A VIF above 10 typically signals high multicollinearity, although even lower values can pose issues depending on the model and context.

```python
def vif(exogs, data):
  vif_dict = {}

  for exog in exogs:
    not_exog = [i for i in exogs if i != exog] #First column is excluded from this list which becomes y in next step
    X, y = data[not_exog], data[exog]

    r_squared = LinearRegression().fit(X,y).score(X,y) 
    #.score(X, y) computes the 𝑅 square value for the fitted model, which measures the proportion of variance in y explained by X.

    # calc the VIF
    vif = 1/(1-r_squared)
    vif_dict[exog] = vif

  return pd.DataFrame({"VIF":vif_dict})
vif_data = vif(X_train_fx.columns, X_train_fx).sort_values(by = 'VIF', ascending = False)
vif_data
```

#### Interpreting VIF Results:

The VIF analysis indicates that several features exhibit infinite (inf) values, pointing to perfect multicollinearity. This occurs when certain features are exact linear combinations of others, making it impossible to estimate their coefficients independently.

**Features with Infinite VIF:**

- **Paint Colors:** All categories for paint color show infinite VIF, indicating perfect collinearity.
- **Vehicle Types:** Categories such as mini-van, hatchback, coupe, etc., have infinite VIF values, signifying perfect multicollinearity.
- **Fuel Types:** Every fuel type category also displays infinite VIF.
- **Transmission Types:** Transmission types exhibit infinite VIF, indicating they are fully collinear with other features.

### Cross Validation and Hyperparameter Tuning

```python
from sklearn.model_selection import cross_val_score, GridSearchCV
# Define the parameter grid for Ridge, Lasso, and Elastic Net
param_grid = {
    'Elastic Net Regression': {'alpha': [0.1, 1.0, 10.0, 100.0]},
    'Ridge': {'alpha': [0.1, 1.0, 10.0, 100.0]},
    'Lasso': {'alpha': [0.01, 0.1, 1.0, 10.0]}
}

# Initialize the models
models = {
    'Elastic Net Regression': ElasticNet(),
    'Ridge': Ridge(),
    'Lasso': Lasso()
}

# Perform cross-validation and hyperparameter tuning
best_models = {}
for name, model in models.items():
    if name in param_grid:
        grid = GridSearchCV(model, param_grid[name], cv=5, scoring='neg_mean_squared_error')
        grid.fit(X_train_fx1, y_train)
        best_models[name] = grid.best_estimator_
        best_rmse = np.sqrt(-grid.best_score_)
        print(f"Best {name}: {grid.best_params_}, RMSE = {best_rmse:.4f}")

print()
print('Evaluate the best models on the test set')
results = {}
for name, model in best_models.items():
    model.fit(X_train_fx, y_train)
    y_pred = model.predict(X_test_fx)
    mse = mean_squared_error(y_test, y_pred)
    rmse = mse ** 0.5
    r2 = r2_score(y_test, y_pred)
    results[name] = {'RMSE': rmse, 'R2': r2}
    print(f"{name}: RMSE = {rmse:.4f}, R2 = {r2:.4f}")
```

Based on the post-tuning metrics, ElasticNet is the best performing model with the lowest RMSE and highest R2 score, making it the preferred choice for predicting used car prices in this context.

#### Best Model: ElasticNet with cross-validation

```python
from sklearn.model_selection import cross_val_score, KFold

# Initialize ElasticNet model
model = ElasticNet()

# Use KFold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation and collect coefficients
coefficients = []
scores = []

for train_index, test_index in kf.split(X_train_fx1):
    X_train_fold, X_test_fold = X_train_fx1.iloc[train_index], X_train_fx1.iloc[test_index]
    y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]
    
    model.fit(X_train_fold, y_train_fold)
    coefficients.append(model.coef_)
    scores.append(model.score(X_test_fold, y_test_fold))

# Convert list of coefficients to a DataFrame
coefficients_df = pd.DataFrame(coefficients, columns=X_train_fx1.columns)

# Evaluate coefficients (e.g., mean and standard deviation across folds)
coefficients_mean = coefficients_df.mean()
coefficients_std = coefficients_df.std()

print("Coefficients mean:")
print(coefficients_mean)
print("\nCoefficients standard deviation:")
print(coefficients_std)

# Evaluate scores
mean_score = np.mean(scores)
std_score = np.std(scores)

print("\nMean cross-validation score:", mean_score)
print("Standard deviation of cross-validation score:", std_score)
```
